## Importação das Bibliotecas 

```python
import pandas as pd 
import numpy as np 
```

```python
energia = pd.read_csv('energia_2022.csv');

energia
```

## 1 - Consumo Total por Município

** 1 - Calcule o consumo total de energia elétrica para cada município somando as colunas "Total (Mwh)". Isso ajudará a identificar os municípios com o maior e menor consumo.

```python
# Calcular o consumo total de energia elétrica para cada município
consumo_total_por_municipio = energia.groupby('Municipios')['Total (Mwh)'].sum().reset_index()

# Classificar os municípios pelo consumo total (do maior para o menor)
consumo_total_por_municipio = consumo_total_por_municipio.sort_values(by='Total (Mwh)', ascending=False)

# Exibir os resultados
display(consumo_total_por_municipio)
```

## 2 - Distribuição de Consumo por Setor

** 1 - Analise a distribuição de consumo de energia elétrica entre os setores residencial, industrial, comercial, rural, poderes públicos, iluminação pública e outros. Isso pode destacar setores que consomem mais energia em cada município.

```python
# Lista de colunas de setores (excluindo a coluna 'Município')
colunas_setores = ['Residencial (Mwh)', 'Industrial (Mwh)', 'Comercial (Mwh)', 'Rural (Mwh)', 'Poderes públicos (Mwh)', 'Iluminação pública (Mwh)', 'Outros (Mwh)']
```

```python
# Converter as colunas de setores para tipos numéricos
for coluna in colunas_setores:
    energia[coluna] = pd.to_numeric(energia[coluna], errors='coerce')
```

```python
# Adicione uma coluna 'Total Setores' que é a soma de todas as colunas de setores
energia['Total Setores'] = energia[colunas_setores].sum(axis=1)
```

```python
# Normalize os dados de setores dividindo pelo total para obter a distribuição percentual
for coluna in colunas_setores:
    energia[coluna + ' (%)'] = (energia[coluna] / energia['Total Setores']) * 100
```

```python
# Exibir o DataFrame resultante com a distribuição percentual de consumo por setor com duas casas decimais
display(energia[['Municipios'] + [coluna + ' (%)' for coluna in colunas_setores]].round(2))
```

## 4 - Identificação de Outliers

** 1 - Procure municípios que apresentem consumo significativamente acima ou abaixo da média para determinados setores. Esses municípios podem ser alvos de investigações adicionais.

```python
# Defina o limiar para identificar outliers, por exemplo, como 2 desvios padrão acima ou abaixo da média
limiar_desvio_padrao = 2
```

```python
# Calcule as estatísticas de resumo para cada setor
setores = ['Residencial (Mwh)', 'Industrial (Mwh)', 'Comercial (Mwh)', 'Rural (Mwh)', 'Poderes públicos (Mwh)', 'Iluminação pública (Mwh)', 'Outros (Mwh)']
```

```python
for setor in setores:
    media = energia[setor].mean()
    desvio_padrao = energia[setor].std()
    limite_superior = media + (limiar_desvio_padrao * desvio_padrao)
    limite_inferior = media - (limiar_desvio_padrao * desvio_padrao)
    
    # Identifique os municípios que estão significativamente acima ou abaixo da média
    outliers_superior = energia[energia[setor] > limite_superior]
    outliers_inferior = energia[energia[setor] < limite_inferior]
    
    print(f"Outliers no setor {setor}:")
    print("Acima da média:")
    print(outliers_superior.round(2))
    print("Abaixo da média:")
    print(outliers_inferior.round(2))
```
